**Yêu cầu 3.2:** Học viên tìm hiểu tập dữ liệu mẫu còn lại trong bộ Datasets của thư viện Sklearn:
1. Cho biết bộ dữ liệu này lưu trữ dữ liệu gì?
2. Có bao nhiêu thuộc tính? bao nhiêu bản ghi?
3. Tập dữ này dùng cho bài toán nào?

```python
#1.Tập dữ liệu Diabetes Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
from sklearn.datasets import load_diabetes
import pandas as pd
# lưu trữ thông tin về bệnh tiểu đường
data = load_diabetes()
df_diabetes = pd.DataFrame(data['data'], columns=data['feature_names'])
df_diabetes['target'] = data['target']
df_diabetes.info()
df_diabetes

df_diabetes.head()
```

```python
df_diabetes.describe()
```

 * tập dữ liệu có 442 bản ghi và có 10 thuộc tính
 * Tập dữ liệu này dùng cho việc kiểm tra xem một người có bị tiểu đường hay không

```python
#2.Tập dữ liệu Digits Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_digits
import pandas as pd

data = load_digits()
df_digits = pd.DataFrame(data['data'], columns=data['feature_names'])
df_digits['target'] = data['target']
df_digits.info()
df_digits
df_digits.head()




```

* Tải và trả về tập dữ liệu chữ số (phân loại).

 * Mỗi điểm dữ liệu là một hình ảnh 8x8 của một chữ số.
 * Có : 10 lớp.
 * Mẫu cho mỗi lớp: 180
 * Tổng số mẫu: 1797
 * Có 64 thuộc tính.

```python
from sklearn.datasets import load_digits
digits = load_digits()
print(digits.data.shape)
# (1797, 64)
import matplotlib.pyplot as plt
plt.gray()
plt.matshow(digits.images[0])

plt.show()
```
